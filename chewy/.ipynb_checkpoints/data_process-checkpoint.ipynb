```python
import re
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
chewy = pd.read_csv('chewy_cat.csv')
```

```python
chewy.head()
```

```python
chewy2 = chewy.copy()
```

```python
chewy2["info_dict"] = chewy2["info_dict"].apply(lambda x : dict(eval(x)) )
chewy3 = chewy2["info_dict"].apply(pd.Series )
```

```python
chewy = pd.concat([chewy, chewy3], axis=1).drop(["info_dict", "Brand", "Breed Size", "Small Pet Type", "Dimensions", "Restricted Availability"], axis=1)
chewy.columns
```

```python
chewy = chewy[["Item Number", "product", "brand", "new_price", "old_price", "num_reviews", "rating", 
               "recommand", "key_benefit", "Weight", "Food Form", "Lifestage", "Food Texture", "Special Diet", 
               "Packaging Type"]]
chewy.head()
```

```python
weight = chewy['Weight'].str.split('\xa0').to_list()
```

```python
#convert ounces to pounds
for i in weight:
    try:
        if i[1]=="ounces":
            i[1]='pounds'
            i[0]=float(i[0])*0.0625
    except:
        i=""
```

```python
chewy['Weight'] = np.array(weight)
```

```python
chewy['Weight'] = chewy['Weight'].apply(lambda  x : x[0] if type(x) == list else x)
```

```python
lifestage = chewy['Lifestage'].apply(pd.Series )
```

```python
lifestage = lifestage.replace('Senior, Adult', 'Adult，Senior').replace('Adult, Kitten', 'Kitten, Adult')
lifestage = lifestage.replace('Adult, Kitten, Senior', 'all stage')
```

```python
chewy['Lifestage'] = lifestage.replace('Adult，Senior', 'Adult, Senior')
```

```python
chewy['Food Form'].value_counts()
```

```python
food_type = chewy['Food Form'].apply(pd.Series )
```

```python
food_type = food_type.replace('Food Topping, Treats', 'Food Topping').replace('Treats, Freeze-Dried', 'Freeze-Dried')
food_type = food_type.replace('Wet Food, Food Topping', 'Wet Food').replace('Freeze-Dried, Food Topping', 'Freeze-Dried')
food_type = food_type.replace('Food Topping, Wet Food, Treats', 'Food Topping').replace('Freeze-Dried, Food Topping, Treats', 'Freeze-Dried')
food_type = food_type.replace('Treats, Freeze-Dried, Food Topping', 'Treats').replace('Wet Food, Food Topping, Supplement', 'Wet Food')
food_type = food_type.replace('Liquid, Food Topping, Treats', 'Food Topping').replace('Air-Dried, Dehydrated', 'Dehydrated')
food_type = food_type.replace('Freeze-Dried, Treats', 'Freeze-Dried').replace('Freeze-Dried, Treats, Food Topping', 'Freeze-Dried')
food_type = food_type.replace('Dry Food, Freeze-Dried', 'Dry Food').replace('Food Topping, Freeze-Dried', 'Food Topping')
food_type = food_type.replace('Food Topping, Freeze-Dried, Powder', 'Food Topping').replace('Treats, Food Topping', 'Food Topping')
food_type = food_type.replace('Treats', 'Food Topping').replace('Food Topping, Powder', 'Food Topping')
food_type = food_type.replace('Dehydrated, Powder, Mix, Treats', 'Dehydrated').replace('Food Topping, Wet Food', 'Food Topping')
food_type = food_type.replace('Dehydrated, Air-Dried', 'Dehydrated').replace('Freeze-Dried, Mix', 'Freeze-Dried')
food_type = food_type.replace( 'Wet Food, Food Topping, Treats', 'Wet Food').replace('Dehydrated, Treats, Food Topping', 'Dehydrated')
food_type = food_type.replace( 'Treats, Dehydrated, Food Topping', 'Dehydrated').replace('Dehydrated, Powder, Mix, Food Topping, Treats', 'Dehydrated')
food_type = food_type.replace('Food Topping, Liquid, Treats', 'Food Topping').replace('Frozen', 'Freeze-Dried')

chewy['Food Form'] = food_type

```

```python
chewy['Food Form'].unique()
```

```python
chewy['Food Texture'].value_counts()
```

```python
Texture = chewy['Food Texture'].apply(pd.Series )
```

```python
Texture = Texture.replace('Chunks in Gravy, Flaked', 'Chunks in Gravy').replace('Pate, Minced', 'Pate')
Texture = Texture.replace('Chunks in Gravy, Pate', 'Chunks in Gravy').replace('Pate, Flaked', 'Pate')
Texture = Texture.replace('Chunks in Gravy, Aspic / Gelee, Minced', 'Chunks in Gravy').replace('Flaked, Chunks in Gravy', 'Flaked')
Texture = Texture.replace('Chunks in Gravy, Aspic / Gelee', 'Chunks in Gravy').replace('Aspic / Gelee, Chunks in Gravy', 'Chunks in Gravy')
Texture = Texture.replace('Pate, Chunks in Gravy', 'Pate').replace('Chunks in Gravy, Grilled', 'Chunks in Gravy')
Texture = Texture.replace('Chunks in Gravy, Minced', 'Chunks in Gravy').replace('Chunks in Gravy, Aspic / Gelee', 'Chunks in Gravy')
Texture = Texture.replace('Aspic / Gelee, Minced', 'Aspic / Gelee').replace('Bits in Broth, Stew', 'Bits in Broth')
Texture = Texture.replace('Chunks in Gravy, Shredded', 'Chunks in Gravy').replace('Chunks in Gravy, Minced, Aspic / Gelee', 'Chunks in Gravy')

Texture = Texture.replace('Minced, Chunks in Gravy', 'Minced').replace('Shredded, Chunks in Gravy', 'Shredded')
Texture = Texture.replace('Aspic / Gelee, Minced', 'Aspic / Gelee').replace('Bits in Broth, Stew', 'Bits in Broth')
Texture = Texture.replace('Shredded, Sliced', 'Shredded').replace('Bits in Broth, Shredded', 'Bits in Broth').replace('Shredded, Pate', 'Shredded')
chewy['Food Texture'] = Texture
```

```python
chewy.to_csv ('chewy.csv', index = None, header=True)
```
